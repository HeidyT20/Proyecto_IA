# Demo - CSP Inventario (Integrado)

Este notebook demuestra el uso del agente y el algoritmo definidos en:
- agent.py
- algorithm.py

Se utiliza Backtracking con restricciones para optimizar inventario.

```python
from src.search_csp.agent import InventoryAgent
from src.search_csp.algorithm import backtracking
```

## Definición del estado inicial

```python
estado_inicial = {
    'inventario': {'A': 50, 'B': 30, 'C': 100},
    'demanda': {'A': 80, 'B': 40, 'C': 90},
    'presupuesto': 10000
}
```

## Parámetros del problema

```python
productos = ['A', 'B', 'C']
costos = {'A': 50, 'B': 100, 'C': 30}
capacidad = 500
```

## Inicialización del agente

```python
agente = InventoryAgent(productos, costos, capacidad)
```

## Ejecución del algoritmo CSP (Backtracking)

```python
resultado = backtracking(estado_inicial, productos, costos, capacidad)

print("=== RESULTADO FINAL ===")
print(resultado)
```

## Evaluación del resultado

```python
if resultado:
    print("\nCosto total:", agente.costo_total(resultado))

    print("\nCumple objetivo:", agente.es_meta(resultado))

    print("\nInventario final:")
    for p in productos:
        print(f"{p}: {resultado['inventario'][p]} / demanda: {resultado['demanda'][p]}")
else:
    print("No se encontró solución")
```

## Análisis del resultado

```python
if resultado:
    for p in productos:
        inv = resultado['inventario'][p]
        dem = resultado['demanda'][p]

        if inv >= dem:
            print(f"{p}: Demanda satisfecha")
        else:
            print(f"{p}: Falta inventario ({dem - inv}) unidades")
```
